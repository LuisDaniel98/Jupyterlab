```python
fi=sp.symbols('fi')
theta=sp.symbols('theta')
p=1
x=p*sp.sin(fi)*sp.cos(theta)
y=p*sp.sin(fi)*sp.sin(theta)
z=p*sp.cos(fi)
F=sp.Matrix([pow(x,2),pow(y,2),pow(z,2)])
print(F)
```

```python
r=sp.Matrix([p*sp.sin(fi)*sp.cos(theta),p*sp.sin(fi)*sp.sin(theta),p*sp.cos(fi)])
```

```python
rFi = sp.diff(r,fi)
rTheta = sp.diff(r,theta)

# Mostrar las derivadas
print(f"Derivada parcial respecto a fi: {rFi}")
print(f"Derivada parcial respecto a theta: {rTheta}")
```

```python
productCruz=rFi.cross(rTheta)
```

```python
dot=F.dot(productCruz)
```

```python
iter1=sp.integrate(dot,(fi,0,np.pi))
print(iter1,p)
```

```python
iter2=sp.integrate(iter1,(theta,0,2*np.pi))
print(iter2)
```
